```python
# IMPORTS ----------------------------------
from astropy.io import fits
from astropy import wcs

import os
import numpy as np
import matplotlib.pyplot as plt
import aplpy
```

```python
target_url = 'https://www2.mpia-hd.mpg.de/THINGS/Data_files/NGC_6946_NA_MOM0_THINGS.FITS'

IMAGE_FITS = fits.open(target_url)
```

```python
IMAGE_FITS[0].header['TELESCOP']
```

```python

fig = plt.figure(facecolor='w', edgecolor='w', frameon=True, figsize=(10,9))
ax = aplpy.FITSFigure(IMAGE_FITS, figure=fig)

ax.show_colorscale(cmap='magma')
ax.add_grid()
ax.grid.show()
ax.set_title(os.path.basename(target_url))

fig.canvas.draw()
```
